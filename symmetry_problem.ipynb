**Symmetry problem and solution using xavier and He**

```python
#Xavier method
import math
fin=10
fout=20
w=2/(fin+fout)
w=math.sqrt(w)
print(w)
```

```python
import torch.nn as nn
layer=nn.Linear(10,20)
w=nn.init.xavier_normal_(layer.weight)
print(w)
```

```python
import torch.nn as nn
layer=nn.Linear(10,20)
w=nn.init.xavier_uniform_(layer.weight)
print(w)
```

```python
#He method
import math
fin=100
w=2/fin
w=math.sqrt(w)
print(w)
```

```python
import torch.nn as nn
layer=nn.Linear(20,10)
w=nn.init.kaiming_normal_(layer.weight)
print(w)
```

```python
import torch.nn as nn
layer=nn.Linear(20,10)
w=nn.init.kaiming_uniform_(layer.weight)
print(w)
```

```python
#diff between Havier and He for Relu
import torch
import torch.nn as nn

torch.manual_seed(0)

def dead_relu_experiment(init_type):
    x = torch.randn(1000, 100)  # input
    dead_counts = []

    for layer_idx in range(50):
        layer = nn.Linear(100, 100)

        if init_type == "xavier":
            nn.init.xavier_normal_(layer.weight)
        else:  # he
            nn.init.kaiming_normal_(layer.weight, nonlinearity="relu")

        x = layer(x)
        x = torch.relu(x)

        # Count dead neurons (output always zero)
        dead_neurons = (x.abs().sum(dim=0) == 0).sum().item()
        dead_counts.append(dead_neurons)

    print(f"\n{init_type.upper()} initialization")
    for i, d in enumerate(dead_counts):
        print(f"Layer {i+1}: dead neurons = {d}/100")

dead_relu_experiment("xavier")
dead_relu_experiment("he")
```
