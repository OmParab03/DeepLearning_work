```python
import numpy as np

x=np.random.randn(100,1)*10
y = 2 * x + 1 + np.random.randn(100, 1) * 2

w=np.random.randn(1,1)*5
b=np.random.randn(1,1)*5

learning_rate=0.05
epoches=50
gradient_m=[]
losses=[]
for i in range(epoches):
  y_pred=x @ w + b
  loss=np.mean((y-y_pred)**2)
  losses.append(loss)

  d_loss_d_y_pred = 2 * (y_pred - y) / len(x)
  grad_w = x.T @ d_loss_d_y_pred
  grad_b = np.sum(d_loss_d_y_pred, axis=0)

  gradient_magnitude = np.sqrt(np.sum(grad_w**2) + np.sum(grad_b**2))
  gradient_m.append(gradient_magnitude)


  w -= learning_rate * grad_w
  b -= learning_rate * grad_b

  if i % 5 == 0 or i == epoches - 1:
        print(f"Epoch {i+1}/{epoches}, Loss: {loss:.4f}, Gradient Magnitude: {gradient_magnitude:.4f}, Weight: {w[0][0]:.4f}, Bias: {b[0][0]:.4f}")
```
